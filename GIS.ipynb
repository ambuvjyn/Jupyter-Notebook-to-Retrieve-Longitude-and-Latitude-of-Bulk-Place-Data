```python
# importing pandas
import pandas as pd

# importing geopy library
from geopy.geocoders import Nominatim

df = pd.read_csv("places.csv")

df
```

```python
# calling the Nominatim tool
loc = Nominatim(user_agent="GetLoc")

for index, row in df.iterrows():
    address = tuple([row["Address"]])
    getLoc = loc.geocode(address)
        
    if getLoc is None:
        message = str(address)
        message += "null\n"
    
    else:
    #message += str(address)
        message = getLoc.address
        message += f";{getLoc.latitude};{getLoc.longitude}\n"

    file_path = f'places_with_coordinates.txt'
    with open(file_path, 'a') as file:
        file.write(message)
```

```python
df_res = pd.read_csv("places_with_coordinates.txt", sep=";", header=None)
```
